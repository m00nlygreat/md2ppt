```python
def printo(obj):
    for attr in dir(obj):
        if attr.startswith("_"):
            continue  # 내부 속성 무시
        try:
            value = getattr(obj, attr)
            if callable(value):
                continue  # 메서드는 제외
            print(f"{attr}: {value}")
        except Exception as e:
            print(f"{attr}: <unreadable attribute> ({e})")
```

```python
from pptx import Presentation
from pptx.util import Inches

# 새 프레젠테이션 생성
prs = Presentation()

# 슬라이드 추가 (타이틀 및 콘텐츠 레이아웃 사용)
slide_layout = prs.slide_layouts[5]  # 제목 없는 콘텐츠 전용 레이아웃
slide = prs.slides.add_slide(slide_layout)

# 제목이 있다면 여기에 추가 가능
title_shape = slide.shapes.title
if title_shape:
    title_shape.text = "테이블 예제"

# 테이블 추가할 위치 및 크기 설정
left = Inches(1)
top = Inches(2)
width = Inches(6)
height = Inches(1.5)

# 행과 열 수 정의
rows = 3
cols = 3

# 테이블 추가
table = slide.shapes.add_table(rows, cols, left, top, width, height).table

# 헤더 내용
table.cell(0, 0).text = '이름'
table.cell(0, 1).text = '나이'
table.cell(0, 2).text = '직업'

# 데이터 행
data = [
    ['홍길동', '30', '개발자'],
    ['김영희', '27', '디자이너인지 개발자인지 기획자인지 모르겠음아이구 힘들다'],
]

for i, row_data in enumerate(data):
    for j, cell_text in enumerate(row_data):
        table.cell(i+1, j).text = cell_text  # i+1: 헤더 다음 행부터

print(table.cell(2,2).span_width)

# 저장
prs.save('table_example.pptx')
```

```python
from pygments import lex
from pygments.lexers import get_lexer_by_name

def highlight_to_dict(code: str):
    """
    code 문자열을 토큰화한 후,
    [{'type': 'Token.Name.Function', 'value': 'foo'}, …] 형태의 리스트 반환
    """
    raw = lex(code, get_lexer_by_name(lang))
    print(raw.__str__())
    return [
        {'type': str(ttype), 'value': value}
        for ttype, value in raw
    ]

# 사용 예
tokens = highlight_to_dict(code)
print(tokens)
```
