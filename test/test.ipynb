```python
layouts = ppt.slide_layouts
for i, l in enumerate(layouts):
    
    print(f"Layout: {i} - {l.name}")
    
    arr = []
    
    slide = ppt.slides.add_slide(l)
    
    for j, p in enumerate(slide.placeholders):
        arr.append([j, p.name, p.placeholder_format.idx, layouts[i].placeholders[j].name, layouts[i].placeholders[j].placeholder_format.idx])
        
    df = pd.DataFrame(arr, columns=["index", "p_name on slide", "idx on slide", "p_name in layout", "idx in layout"])
    display(df)
```

```python
code = """def date():
    # 패턴 문자열로 정의합니다.
    # pattern = r'^\[(?P<key>[^\]]+)\]:\s*#\s*\((?P<value>[^)]+)\)\s*$'
    pattern = r'^\[(?P<key>[^\]]+)\]:\s*#\s*\((?P<value>.+)\)\s*$'
    # 콜백 함수는 (self, m, state)를 받고, 토큰을 상태에 추가한 후 새 커서 위치를 정수로 반환해야 합니다.
    def parse_comment_block(self, m, state):
        token = {
            'type': 'comment_block',
            'key': m.group('key'),
            'value': m.group('value'),
            'raw': m.group(0).replace('\n','')
        }
        state.append_token(token)
        return m.end()
    
    # 'ref_link'보다 먼저 처리되도록 등록합니다.
    md.block.register('comment_block', pattern, parse_comment_block, before='ref_link')
    return md
"""
```

```python
from pygments import lex
from pygments.lexers import PythonLexer

def highlight_to_dict(code: str):
    """
    code 문자열을 토큰화한 후,
    [{'type': 'Token.Name.Function', 'value': 'foo'}, …] 형태의 리스트 반환
    """
    raw = lex(code, PythonLexer())
    return [
        {'type': str(ttype), 'value': value}
        for ttype, value in raw
    ]

# 사용 예
tokens = highlight_to_dict(code)
print(tokens)
```
